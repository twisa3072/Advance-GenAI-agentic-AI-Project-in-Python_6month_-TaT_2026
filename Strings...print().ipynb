# .Format Method

```python
num1 = 40
num2 = 50
add=num1 + num2
print('The addition of {} and {} is = {}'.format(num1,num2,add))
```

```python
num1 = 40
num2 = 50
sub=num1 - num2
print('The subtraction of {} and {} is = {}'.format(num1,num2,sub))
```
