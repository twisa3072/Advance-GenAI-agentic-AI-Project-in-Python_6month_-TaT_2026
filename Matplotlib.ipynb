State Machine Interface

```python
x = np.linspace(1,21,5)
```

```python
plt.plot(x, x, label='linear')
plt.plot(x,x**2, label='quadratic')
plt.plot(x,x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")
plt.legend()

plt.show()
```

Error Bar Chart

```python
x9 = np.arange(0,4,0.2)
y9= np.exp(-x9)
e1 = 0.1*np.abs(np.random.randn(len(y9)))
plt.errorbar(x9,y9,yerr=e1,fmt='.-')
plt.show()
```

Stacked Bar Chart

```python
A=[15,30,45,22]
B=[15,25,50,20]
z2=range(4)
plt.bar(z2,A,color='blue')
plt.bar(z2,B,color='Green',bottom=A)
plt.show()
```

Pie Chart

```python
plt.figure(figsize=(7,7))
x10=[35,25,20,20]
labels=['Computers','Electronics','Electrical','Civil']
plt.pie(x10,labels=labels, autopct=lambda p:f'{int(round(p*sum(x10)/100))}')
plt.show()
```
